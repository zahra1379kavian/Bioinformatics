```python
file = open('rosalind.txt','r')
Lines = file.readlines()
lines = []
seq = ""


for line in Lines:
    if line.find('>') == 0:
        lines.append(seq)
        seq = ""
    if line.find('>') != 0:
            seq = seq + line.strip('\n')

lines.append(seq)
lines.pop(0)
```

```python
def score(x):
    score = 0
    for i in range(len(x)):
        for j in range(i+1, len(x)):
            if x[i] != x[j]:
                score -= 1
    return score
```
