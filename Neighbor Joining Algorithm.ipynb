```python
import numpy as np
```

```python
f = open('rosalind.txt','r')
data = []

for line in f:
    data.append(line.strip())
    
n = int(data[0])
disMatrix = np.zeros((n,n))

for i in range(n):
    a = data[i+1].split()
    for j in range(n):
        disMatrix[i,j] = int(a[j])
```

```python
def print_adj(adj):
    for i, node in enumerate(adj):
        for d,w in node:
            print(str(i)+'->'+str(d)+':'+'%0.3f'%w)
```

```python
clusters = [i for i in range(n)]
adj = [[] for i in range(n)]

while True:
    if len(disMatrix)==2:
        adj[len(adj)-1].append((len(adj)-2, disMatrix[0][1]))
        adj[len(adj)-2].append((len(adj)-1, disMatrix[0][1]))
        print_adj(adj)
        break
    
    row_sum = np.sum(disMatrix,axis=0)
    D1 = (n-2)*disMatrix-row_sum-row_sum.reshape((n,1))
    np.fill_diagonal(D1, 0.)
    
    index = np.argmin(D1)
    i = index // n
    j = index % n
    
    delta = (row_sum[i]-row_sum[j])/(n-2)
    limb_lengthi = (disMatrix[i,j]+delta)/2
    limb_lengthj = (disMatrix[i,j]-delta)/2
    
    D_new = (disMatrix[i,:]+disMatrix[j,:]-disMatrix[i,j])/2
    disMatrix = np.insert(disMatrix,n,D_new,axis=0)
    D_new = np.insert(D_new,n,0.,axis=0)
    disMatrix = np.insert(disMatrix,n,D_new,axis=1)
    disMatrix = np.delete(disMatrix,[i,j],axis=0)
    disMatrix = np.delete(disMatrix,[i,j],axis=1)
    
    L = len(adj)
    adj.append([])
    adj[L].append((clusters[i],limb_lengthi))
    adj[clusters[i]].append((L,limb_lengthi))
    adj[L].append((clusters[j],limb_lengthj))
    adj[clusters[j]].append((L,limb_lengthj))
    
    if i < j:
        del clusters[j]
        del clusters[i]
    else:
        del clusters[i]
        del clusters[j]
        
    clusters.append(L)
    
    n -= 1
```
