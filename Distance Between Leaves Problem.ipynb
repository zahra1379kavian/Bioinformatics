```python
import queue
```

```python
f = open('rosalind.txt','r')
data = []

for line in f:
    data.append(line.strip())

n = int(data[0])
adjDict = dict()
for d in data[1:]:
    d = d.split('->')
    d1 = d[1].split(':')
    if not int(d[0]) in adjDict:
        adjDict[int(d[0])] = []
    adjDict[int(d[0])].append((int(d1[0]),int(d1[1])))
print(adjDict)
```

```python
distMatrix = [[0]*n for _ in range(n)]

for i in range(n):
    dist = dict()
    
    dist[i] = 0
    q = queue.Queue()
    q.put(i)
    while not q.empty():
        currNode = q.get()
        for node, weight in adjDict[currNode]:
            if not node in dist:
                dist[node] = dist[currNode] + weight
                if node < n:
                    distMatrix[i][node] = dist[node]
                q.put(node)
```

```python
for d in distMatrix:
    print(' '.join([str(i) for i in d]))
```
