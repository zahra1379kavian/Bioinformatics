```python
import numpy as np
from copy import deepcopy
```

```python
f = open('rosalind.txt','r')
data = []

for line in f:
    data.append(line.strip())
    
n = int(data[0])
disMatrix = np.zeros((n,n))

for i in range(n):
    a = data[i+1].split()
    for j in range(n):
        if i==j:
            disMatrix[i,j] = np.inf
        else:
            disMatrix[i,j] = int(a[j])
```

```python
clusters = [[i,1] for i in range(n)]
adj = [[] for _ in range(n)]
Age = [0. for _ in range(n)]
flag = True

while 1:
    index = np.argmin(disMatrix)
    i = index // len(disMatrix)
    j = index % len(disMatrix)
    
    L = len(adj)
    adj.append([])
    
    C_new = [L, clusters[i][1]+clusters[j][1]]
    adj[L].append(clusters[i][0])
    adj[L].append(clusters[j][0])
    adj[clusters[i][0]].append(L)
    adj[clusters[j][0]].append(L)
    
    
    Age.append(disMatrix[i,j]/2)
    
    if len(disMatrix) == 2:
        break
    
    d_new = (disMatrix[i,:]*clusters[i][1]+disMatrix[j,:]*clusters[j][1])/(clusters[i][1]+clusters[j][1])
    d_new = np.delete(d_new,[i,j],0)
    disMatrix = np.delete(disMatrix,[i,j],0)
    disMatrix = np.delete(disMatrix,[i,j],1)
    disMatrix = np.insert(disMatrix,len(disMatrix),d_new,0)
    d_new = np.insert(d_new,len(d_new),np.inf,0)
    disMatrix = np.insert(disMatrix,len(disMatrix)-1,d_new,1)
    
    if i<j:
        del clusters[j]
        del clusters[i]
    else:
        del clusters[i]
        del clusters[j]
        
    clusters.append(C_new)
    
    
    
adj_new = deepcopy(adj)
for i, nodes in enumerate(adj):
    for j, v in enumerate(nodes):
        adj_new[i][j] = (v, abs(Age[i]-Age[v]))
```

```python
for i, nodes in enumerate(adj_new):
    for  d,w in nodes:
        print(str(i)+'->'+str(d)+':'+str('%0.3f'%w))
```
