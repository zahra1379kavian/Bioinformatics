```python
import numpy as np

file = open('rosalind.txt','r')
Lines = file.readlines()
lines = []
seq = ""


for line in Lines:
    if line.find('>') == 0:
        lines.append(seq)
        seq = ""
    if line.find('>') != 0:
            seq = seq + line.strip('\n')

lines.append(seq)
lines.pop(0)
```

```python
def reverse_com(input_string):
    seq = (input_string.replace("A", "t").replace("C", "g").replace("T", "a").replace("G", "c")).upper()
    out_string = seq[::-1]
    return out_string
```

```python
def check_seq(s,checked_list,lines):
    #S = [s,reverse_com(s)]
    if s in checked_list:
        index = lines.index(s)
    elif reverse_com(s) in checked_list:
        index = lines.index(reverse_com(s))    
    else:
        index = -1
    return index
```

```python
def hamming(seq1,seq2):
    mutations = 0
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]:
            mutations += 1
    return mutations
```

```python
checked_list = [lines[0]]
number = np.ones((1,len(lines)))


for i in range(1,len(lines)):
    #print(lines[i])
    index = check_seq(lines[i],checked_list,lines)
    #print(index)
    if index != -1:
        number[0,index] += 1
        number[0,i] += 1
    else:
        checked_list.append(lines[i])
    
correct = []
incorrect = []

for i in range(len(lines)):
    if number[0,i] == 1:
        incorrect.append(lines[i])
    else:
        correct.append(lines[i])
        correct.append(reverse_com(lines[i]))

corrected = []
for i in range(len(incorrect)):
    for j in range(len(correct)):
        if hamming(incorrect[i],correct[j]) == 1:
            corrected.append(correct[j])
            break

for i in range(len(incorrect)):
    print(incorrect[i]+'->'+corrected[i])
```
