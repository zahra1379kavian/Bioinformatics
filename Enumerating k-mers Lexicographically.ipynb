```python
import itertools

with open('rosalind_lexf.txt') as file:
    lines = [line.rstrip() for line in file]

seq = lines[0].replace(" ", "")
n = int(lines[1])

ord_seq = list(seq.encode('ascii'))
ord_seq.sort()

permu = [p for p in itertools.product(ord_seq, repeat=n)]

for j in list(permu):
    print(''.join(chr(i) for i in j))

```
