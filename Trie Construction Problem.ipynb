```python
file = open('rosalind_ba9a.txt','r')
seq = []

for line in file:
    seq.append(line.strip())

#del seq[-1]
```

```python
class Trie:
    def __init__(self):
        self.Trie_construction(seq)
    
    def Trie_construction(self, seq):
        trie = dict() 
        trie[0] =  dict()
        node = 1

        for pattern in seq:
            root = 0
            for letter in pattern:
                if letter in trie[root]:
                    root = trie[root][letter]
                else:
                    trie[root][letter] = node
                    trie[node] = dict()
                    root = node
                    node += 1
        return trie
```

```python
t = Trie()
trie = t.Trie_construction(seq)
```

```python
for node in trie:
    for child in trie[node]:
        print('{}->{}:{}'.format(node,trie[node][child],child))
```
