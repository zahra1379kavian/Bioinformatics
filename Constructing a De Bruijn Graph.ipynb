```python
def reverse_com(input_string):
    seq = (input_string.replace("A", "t").replace("C", "g").replace("T", "a").replace("G", "c")).upper()
    out_string = seq[::-1]
    return out_string
```

```python
import numpy as np

file = open('rosalind.txt','r')
Lines = file.readlines()
lines = []
seq = ""

for seq in Lines:
    lines.append(seq.strip('\n'))
```

```python
S = []
[S.append(x) for x in lines if x not in S]
```

```python
Src = []
[Src.append(reverse_com(x)) for x in S]


[S.append(x) for x in Src if x not in S]
```
